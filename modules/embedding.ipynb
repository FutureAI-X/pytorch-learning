```python
import torch
import torch.nn as nn
```

# 一 基本用法

定义Embedding，词表大小为10，每个词使用一个5维的向量表示

```python
embedding = nn.Embedding(num_embeddings=10, embedding_dim=5)
embedding.weight.data
```

```python
input = torch.tensor(0)
output = embedding(input)
print(input.shape)
print(output)
print(output.shape)
```

```python
input = torch.tensor([0])
print(input.shape)
output = embedding(input)
print(output)
print(output.shape)
```

```python
input = torch.tensor([0, 1, 2])
print(input.shape)
output = embedding(input)
print(output)
print(output.shape)
```

```python
input = torch.tensor([[0, 1, 2]])
print(input.shape)
output = embedding(input)
print(output)
print(output.shape)
```

```python
input = torch.tensor([[0, 1, 2],[3, 4, 5]])
print(input.shape)
output = embedding(input)
print(output)
print(output.shape)
```

# 二 Padding

```python
embedding = nn.Embedding(num_embeddings=10, embedding_dim=5, padding_idx=0)
embedding.weight.data
```

```python
input = torch.tensor([[0, 1, 2], [0, 0, 3]])
print(input.shape)
output = embedding(input)
print(output)
print(output.shape)
```
